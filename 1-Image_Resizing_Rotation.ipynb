### Image Resizing

## 1. Resizing with Custom Height and Width
## 2. Using Scaling Factor
* Scaling Factor or Scale Factor is usually a number that scales or multiplies some quantity, in our case the width and height of the image. It helps keep the aspect ratio intact and preserves the display quality. So the image does not .
appear distorted, while you are upscaling or downscaling it

## 3. Resizing with Interpolation Methods
* Inter Area - for reducing the size of an image
* Inter Cubic -  slow
* Inter Linear -  For Zooming purposes
* Inter Nearest - Uses nearest neighbors concept

```python
def resize(image, width, height):

  print("\n\n*********************USING CUSTOM WIDTH AND HEIGHT******************")

  print()
  resize_points = (width, height)
  scaling = cv2.resize(image, resize_points)
  print("\nResized Image")
  print(scaling.shape)
  cv2_imshow(scaling)

  print("\n\n\n***************USING SCALE FACTOR*********************")
  scale_up_width = float(input("Scale-up Width: "))
  #scale_up_height = round(scale_up_width/aspect, 2)
  print("Scale-up Height: ", scale_up_width)
  scale_down = float(input("Scale Down Value: "))
  print("\n")

  print("Original Image")
  print(image.shape)
  cv2_imshow(image)

  print()

  #fy_value= round(scale_down/aspect,2)
  scaled_image1 = cv2.resize(image,None, fx = scale_down, fy = scale_down) #None -Desired size of output image
  print("Shape: ",scaled_image1.shape, "\nX-Axis:", scale_down, "\nY-axis:", scale_down)
  cv2_imshow(scaled_image1)


  scaled_image2 = cv2.resize(image, None, fx =scale_up_width , fy =scale_up_width)
  print()
  print("Shape: ",scaled_image2.shape, "\nX-Axis:", scale_up_width, "\nY-axis:", scale_up_width)
  cv2_imshow(scaled_image2)
  print()

  print("\n\n\n******************INTERPOLATION METHODS************************")
  print("Original Image")
  print(image.shape)
  cv2_imshow(image)
  print()

  fx_area= 0.75
  fy_area = 0.75
  small = cv2.resize(image, None, interpolation = cv2.INTER_AREA, fx= fx_area,fy =fy_area)
  print("\nInter-Area", "\nShape:",small.shape, "\nX-axis:",fx_area,"Y-axis:",fy_area)
  cv2_imshow(small)

  print()
  fx_near= 0.75
  fy_near =0.75
  near = cv2.resize(image, None, interpolation = cv2.INTER_NEAREST, fx = fx_near, fy = fy_near)
  print("Inter-Nearest","\nShape:",near.shape, "\nX-axis:",fx_near,"Y-axis:",fy_near)
  cv2_imshow(near)

#big = cv2.resize(img1, None, interpolation  = cv2.INTER_CUBIC, fx =0.5, fy=0.5)
#zoom = cv2.resize(img1, (120,120), interpolation =cv2.INTER_LINEAR)
# print("\nInter-Cubic")
# cv2_imshow(big)
# print("\nInter-Linear")
# cv2_imshow(zoom)


print("Original Image")
("Width:",img1.shape[1], "\nHeight: ", img1.shape[0])
aspect=round(img1.shape[1]/img1.shape[0],2)
print("Aspect Ratio: ", aspect)
cv2_imshow(img1)
print()
```

# Rotation

```python

img1 = cv2.imread("/content/gdrive/MyDrive/Colab Notebooks/CV_IMAGES/NAMO1.jfif", cv2.IMREAD_COLOR)
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

degree_90 = cv2.rotate(img1, cv2.ROTATE_90_CLOCKWISE)
degree_180 = cv2.rotate(img1, cv2.ROTATE_180)
degree_90_counter = cv2.rotate(img1, cv2.ROTATE_90_COUNTERCLOCKWISE)
try:
  angle = int(input("Enter the angle: "))
  user = imutils.rotate_bound(img1, angle=angle)


  print()
  titles = ["Original Image", "90` Rotation", "180` Rotation", "90`_CounterClock", angle]
  rotations = [img1, degree_90,degree_180,degree_90_counter, user]
  plt.figure(figsize=(10,10))
  for i in range(5):
    plt.subplot(2,3,i+1)
    image =  rotations[i]
    plt.imshow(image)
    plt.title(titles[i], fontdict={'fontsize': 15})
  plt.show()

except:
  print("Invalid Rotation Angle")

```
