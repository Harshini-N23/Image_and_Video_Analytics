```python
import cv2
import os
from google.colab.patches import cv2_imshow
import datetime
import glob
import math
```

```python
def frames(vid):
  i = 0
  total_frames= 0
  while(True):
      ret,frame = vid.read()
      if ret:
          cv2.imwrite("/content/drive/MyDrive/Colab Notebooks/CV_IMAGES/output_images/image"+str(i)+".jpg", frame)
          #cv2_imshow(frame)
          total_frames += 1
          i=i+1
      else:
          break
  # Frames per second
  frames_per_sec = math.ceil(vid.get(cv2.CAP_PROP_FPS))
  print('frames per second =',frames_per_sec)
  # Total Frames
  print("Total Frames: ",total_frames )
  # frames = vid.get(cv2.CAP_PROP_FRAME_COUNT) - inbuilt function to get the total_frame_count
  # duration of the video
  seconds = round(total_frames / frames_per_sec)
  video_time = datetime.timedelta(seconds=seconds)
  print(f"duration in seconds: {seconds}")
  print(f"video time: {video_time}")
  vid.release()

def video_from_frames(img_array,size):
  original_video = cv2.VideoWriter('/content/drive/MyDrive/Colab Notebooks/CV_IMAGES/Videos/output_video.avi',cv2.VideoWriter_fourcc(*'DIVX'), 7, size)
  for i in range(len(img_array)):
    original_video.write(img_array[i])
  original_video.release()


def slow_video(img_array,size):
  slow_video =cv2.VideoWriter("/content/drive/MyDrive/Colab Notebooks/CV_IMAGES/Videos/slow_output_video.avi", cv2.VideoWriter_fourcc(*"DIVX"), 2, size)
  for i in range(len(img_array)):
    slow_video.write(img_array[i])


def fast_video(img_array,size):
  fast_video =cv2.VideoWriter("/content/drive/MyDrive/Colab Notebooks/CV_IMAGES/Videos/fast_output_video.avi", cv2.VideoWriter_fourcc(*"DIVX"), 13, size)
  for i in range(len(img_array)):
    fast_video.write(img_array[i])


def reverse():
  video = cv2.VideoCapture("/content/drive/MyDrive/Colab Notebooks/CV_IMAGES/Videos/original_vid.gif")
  frame_list = []
  while(True):
      ret,frame = video.read()
      if ret:
        frame_list.append(frame)
      else:
        break
  frame_list.pop()
  frame_list.reverse()
  reverse_original_video = cv2.VideoWriter('/content/drive/MyDrive/Colab Notebooks/CV_IMAGES/Videos/reversed_output_video.avi',cv2.VideoWriter_fourcc(*'DIVX'), 7, (376,240))





  for i in range(len(frame_list)):
    reverse_original_video.write(frame_list[i])
  reverse_original_video.release()

def gray_video():
  video = cv2.VideoCapture("/content/drive/MyDrive/Colab Notebooks/CV_IMAGES/Videos/original_vid.gif")
  i = 0
  total_frames= 0
  while(True):
      ret,frame = video.read()
      if ret:
          gray = cv2.cvtColor(frame,cv2.COLOR_RGB2GRAY)
          cv2.imwrite("/content/drive/MyDrive/Colab Notebooks/CV_IMAGES/output_images/gray_images/gray_image"+str(i)+".jpg", gray)
          #cv2_imshow(gray)
          total_frames += 1
          i=i+1
      else:
          break

  gray_img_array = []
  for filename in glob.glob('/content/drive/MyDrive/Colab Notebooks/CV_IMAGES/output_images/gray_images/*.jpg'):
      gray_img = cv2.imread(filename)
      height, width, layers = gray_img.shape
      size = (width,height)
      gray_img_array.append(gray_img)

  gray_video = cv2.VideoWriter('/content/drive/MyDrive/Colab Notebooks/CV_IMAGES/Videos/gray_video.avi',cv2.VideoWriter_fourcc(*'DIVX'), 7, (376,240))
  for i in range(len(gray_img_array)):
    gray_video.write(gray_img_array[i])

  gray_video.release()

```

```python
def main():
  print("**Video Analysis**")

  print("Extracting Frames from Videos...")
  vid = cv2.VideoCapture("/content/drive/MyDrive/Colab Notebooks/CV_IMAGES/Videos/original_vid.gif")
  frames(vid)
  print("\nFrames Extracted")
  while True:
    print("\n1. Creating Video from frames\n2. Slow Motion Video\n3. Fast Motion Video\n4.Reversed Video\n5.Gray Scale Video\n6.Exit")
    choice = int(input("Enter your choice: "))
    if choice ==1:

      print("\n\nCoverting Frames to Videos..")
      img_array = []
      for filename in glob.glob('/content/drive/MyDrive/Colab Notebooks/CV_IMAGES/output_images/*.jpg'):
        img = cv2.imread(filename)
        height, width, layers = img.shape
        size = (width,height)
        img_array.append(img)
        video_from_frames(img_array,size)


    elif choice ==2:
      slow_video(img_array,size)
      print("\nSlow Motion Video Created.")
    elif choice ==3:
      fast_video(img_array, size)
      print("\nFast Motion Video Created.")


    elif choice ==4:
      reverse()
      print("\nReverse Motion Video Created.")


    elif choice==5:
      gray_video()
      print("\n Video Created")
    elif choice ==6:
      print("Done")
      break
    else:
      print("Invalid Choice:")
      break





main()

```
